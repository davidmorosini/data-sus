# Merge Database

Nesta fase vamos capturar todo o banco de dados original e unificar os mesmos em arquivos csv, mergeados


```python
list_raw_files = os.listdir(raw_database_dir)
list_raw_files = [d for d in list_raw_files if d.endswith(".csv")]
list_raw_files.sort()

print(list_raw_files)
```

### Executando o merge de cada um dos arquivos CSV do raw database

```python
merged_dir = ile_dir = os.path.join(merged_database_dir, "SIH_2016.csv")
for f in list_raw_files:
    file_dir = os.path.join(raw_database_dir, f)
    print(f"Processing file: {file_dir}")
    
    # Lendo o arquivo csv para utilizar as colunas e complementar caso esteja faltando alguma
    df = pd.read_csv(file_dir)
    aux_columns = list(df.columns)
    aux_columns.extend(usecols)
    aux_columns = get_list_unique_values(aux_columns)
    
    aux_df = pd.DataFrame(columns=aux_columns)
    aux_df = aux_df.append(df, ignore_index=True)
    
    # Dataframe com todas as colunas de interesse selecionadas
    final_df = pd.DataFrame(aux_df, columns=usecols)
    
    del df
    del aux_df
    
    # Filtrando pela coluna DIAG_PRINC com os valores de interesse
    filter_df = final_df.loc[
        final_df["DIAG_PRINC"].isin(rational)
    ]
    
    out_dir = os.path.join(splited_database_dir, f)
    print(f"\tSave result in {out_dir}")
    filter_df.to_csv(out_dir, index=False)
    del filter_df
```

## Consolidando a base final

```python
list_files_splited = os.listdir(splited_database_dir)
list_files_splited = [d for d in list_files_splited if d.endswith(".csv")]
list_files_splited.sort()

print(list_files_splited)
```

```python
df_final = pd.DataFrame(columns=usecols)

for f in list_files_splited:
    file_dir = os.path.join(splited_database_dir, f)
    print(f"Processing file: {f}")
    df = pd.read_csv(file_dir, usecols=usecols)
    
    df_final = df_final.append(df, ignore_index=True)
    print("\tMerged: OK")
    del df

final_dir = os.path.join(final_database_dir, "SIH_CONSOLID.csv")
df_final.to_csv(final_dir, index=False)
```
